```python
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt
import pyccl as ccl
```

```python
def alpha_t_plus_GW(aM,A):
    root = (100*aM**2 - 400*aM + 400)*A**4 - 180*aM*A**2 + 360*aM +81
    nom = (10*aM+20)*A**2 - (20*aM + 9) + np.sqrt(root)
    denom = 40*(2*A**4 -(2+aM)*A**2 + aM)
    return nom / denom

def alpha_t_minus_GW(aM,A):
    root = (100*aM**2 - 400*aM + 400)*A**4 - 180*aM*A**2 + 360*aM +81
    nom = (10*aM+20)*A**2 - (20*aM + 9) - np.sqrt(root)
    denom = 40*(2*A**4 -(2+aM)*A**2 + aM)
    return nom / denom

```
